## Plots using MatplotLib for Data Visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,10,12,16])
plt.plot(x)
# plt.show()
x2 = np.array([5,4,3,7,11,13])
plt.plot(x2)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np. array([10,20,50,80,90])

plt.plot(x,y)
plt.plot(x,x**2)
plt.show()

plt.plot(x,np.sin(x))
```

```python
plt.scatter(x,y,color='red',marker='^')
plt.title("Stock Data")
plt.xlabel("days")
plt.ylabel("Price")
plt.show()
```

```python
x = np.random.random((100,2))

for ix in range(x.shape[0]):
    plt.scatter(x[ix,0], x[ix,1], c='green', marker='x')
    
plt.show()
```

### More on plotting

```python
x = np.array([1,2,3,4,5])
y = np.array([10,20,50,80,90])

plt.plot(x,x**2, label='2018')
plt.plot(x,x**3, label='2017')
plt.xlabel("No of Days")
plt.ylabel("Prices")
plt.title("Stock price Comparission")
plt.legend()
plt.show()
```
