## Plots using MatplotLib for Data Visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,10,12,16])
plt.plot(x)
# plt.show()
x2 = np.array([5,4,3,7,11,13])
plt.plot(x2)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np. array([10,20,50,80,90])

plt.plot(x,y)
plt.plot(x,x**2)
plt.show()

plt.plot(x,np.sin(x))
```

```python
plt.scatter(x,y,color='red',marker='^')
plt.title("Stock Data")
plt.xlabel("days")
plt.ylabel("Price")
plt.show()
```

```python
x = np.random.random((100,2))

for ix in range(x.shape[0]):
    plt.scatter(x[ix,0], x[ix,1], c='green', marker='x')
    
plt.show()
```

### More on plotting

```python
x = np.array([1,2,3,4,5])
y = np.array([10,20,50,80,90])

plt.plot(x,x**2, label='2018')
plt.plot(x,x**3, label='2017')
plt.xlabel("No of Days")
plt.ylabel("Prices")
plt.title("Stock price Comparission")
plt.legend()
plt.show()
```

# Bar Chart

```python
numbers = np.random.randint(0,10,5)
indices = np.arange(5)

numbers2 = np.random.randint(0,10,5)
indices2 = np.arange(5)

print(numbers)
print(indices)
```

```python
plt.bar(indices-0.125,numbers,0.25,color='green',label='GOLD')
plt.bar(indices2+0.125,numbers2,0.25,color='orange',label='SILVER')
plt.legend()
plt.show()
```

# Pie Chart

```python
labels = ['Algebra','Calculus','Coding','Algorithms']
values = [90,70,60,45]

plt.pie(values,labels=labels,radius=1,explode=(0,0,1,0),shadow=True)
plt.show()
```

# Histogram and Normal Distribution

### Normal Distribution (1-D)

```python
u = 0
sigma = 1 
# such a distribution is also called standard normal distribution

u=5
sigma = 0.2

vals = u + sigma*np.random.randn(200)
print(vals.shape)

plt.hist(vals,10)
plt.show()
```

```python
x = vals
y = np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

## Multivariate Normal Distribution

```python
#BiVariate
mean = np.array([0.0,0.0])
cov = np.array([[.3,0],[0,1]])


mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,0.2],[0.2,1.1]])

dist = np.random.multivariate_normal(mean,cov,500)
dist2 = np.random.multivariate_normal(mean2,cov2,500)
print(dist.shape)

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
